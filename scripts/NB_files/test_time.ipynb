```python
import numpy as np 
```

```python
def make_f_handle(center, sigma, A, R=1.0):
    x0, y0, z0 = center

    def f_handle(x, y, z):
        dot = x * x0 + y * y0 + z * z0
        cos_alpha = np.clip(dot / R**2, -1.0, 1.0)
        alpha = np.arccos(cos_alpha)
        return A * np.exp(-(alpha**2) / (2 * sigma**2))

    return f_handle
```

```python
fi = make_f_handle(np.array([1,1,1]), 0.01, 1)
```
